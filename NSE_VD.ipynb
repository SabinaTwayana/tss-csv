```python
import numpy as np
import pandas as pd

def nse(simulated, observed):
  """Calculate the Nash-Sutcliffe efficiency.

  Args:
    simulated: The simulated discharge data.
    observed: The observed discharge data.

  Returns:
    The NSE value.
  """
  sum_squared_error = np.sum((simulated - observed)**2)
  sum_squared_residual = np.sum((observed - np.mean(observed))**2)
  nse = 1 - sum_squared_error / sum_squared_residual
  # print("mean",np.mean(observed))
  # print("sum:",sum_squared_error)
  # print("sum2",sum_squared_residual)
  return nse


def volume_difference(simulated, observed):
  """Calculate the volume difference between simulated and observed discharge.

  Args:
    simulated: The simulated discharge data.
    observed: The observed discharge data.

  Returns:
    The volume difference.
  """
  simulated_volume = np.sum(simulated)
  observed_volume= np.sum(observed)
  volume_difference = (simulated_volume - observed_volume)/ simulated_volume*100
  return volume_difference

if __name__ == "__main__":
  # Load the simulated and observed discharge data.
  simulated = pd.read_csv(r"D:\GDM_MODEL_1\GDM Model-DISTRIBUTION PACK\GDM Model\calibrationandvalidataion\validation\simulated_data_2009-2015.csv")
  observed = pd.read_csv(r"D:\GDM_MODEL_1\GDM Model-DISTRIBUTION PACK\GDM Model\calibrationandvalidataion\validation\observed_data_2009-2015.csv")
  # print("Simulated Columns:", simulated.columns)
  # print("Observed Columns:", observed.columns)
  # Calculate the NSE and volume difference.
  # nse_value1= nse(simulated["265"], observed["265"])
  # volume_difference1 = volume_difference(simulated["265"], observed["265"])
  nse_value2= nse(simulated["sim"], observed["obs"])
  volume_difference2 = volume_difference(simulated["sim"], observed["obs"])
  # nse_value3= nse(simulated["270"], observed["270"])
  # volume_difference3 = volume_difference(simulated["270"], observed["270"])

  # Print the results.
  # print("NSE1:", nse_value1)
  print("NSE2:",round(nse_value2,3))
  # print("NSE3:", nse_value3)
  # print("Volume difference1:", volume_difference1)
  print("Volume difference2:", volume_difference2)
  # print("Volume difference3:", volume_difference3)
```
